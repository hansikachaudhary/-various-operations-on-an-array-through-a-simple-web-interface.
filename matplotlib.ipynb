```python
import matplotlib.pyplot as plt
import numpy as np

# Define x, y, and z
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# Plot the sine wave
plt.plot(x, y)
plt.title('sine wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()

# Plot the cosine wave (if needed)
plt.plot(x, z)
plt.title('cosine wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show
```

BAR PLOT

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['english','french','latin', 'german' , 'japaneese']
people=[100,20,30,40,50]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('PEOPLE')
plt.show()
```

PIE CHART

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['english','french','latin', 'german' , 'japaneese']
people=[100,20,30,40,50]
ax.pie(people,labels=languages,autopct='%1.1f')
```

SCATTER PLOT


```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.scatter(x,y)
plt.show()
```

3D SCATTER PLOT

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greens')
plt.show
```
